# Apply Token Replacement to Transaction Descriptions

This notebook demonstrates how to apply the finalized `replace_tokens()` function from `preprocess.py` to transform transaction descriptions into a more general, model-friendly format.


```python
import sys
import os
import pandas as pd

sys.path.append(os.path.abspath('..'))

from src.preprocess import replace_tokens

df = pd.read_csv("../data/transactions.csv")
df[['description', 'label']].sample(20, random_state=42)
```

###  Applying the Tokenizer

```python
df['description'] = df.apply(lambda row: replace_tokens(row['description'], row['label']), axis=1)
df[['description', 'label']].sample(20, random_state=42)
```

```python
df.to_csv('../data/transactions_tokenized.csv', index=False)
```
